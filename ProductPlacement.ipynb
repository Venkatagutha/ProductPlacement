```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header= None)
```

```python
# aprior model expects a list of lists in a string format

tran = []
for i in range(0,7501):
    tran.append([str(df.values[i,j]) for j in range (0,20)])
    
```

```python
tran[0]
```

```python
from apyori import apriori # download the apyori library from git and save in the current working dir
rules = apriori(tran,min_support= 0.003, min_confidence=0.2 , min_lift = 3 , min_length = 2)
#Support(A) = (Transactions containing (A))/(Total Transactions) // 3*7/7501
#Confidence(A→B) = (Transactions containing both (A and B))/(Transactions containing A) //
#Lift(A→B) = (Confidence (A→B))/(Support (B)) // 
```

```python
results = list(rules)
```

```python
r = []
for i in results:
    y = print(i)
    r.append(y)
```
